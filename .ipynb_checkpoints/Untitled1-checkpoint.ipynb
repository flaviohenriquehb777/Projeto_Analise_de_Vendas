```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df_vendas['VendaQtd'], bins=30, kde=True)
plt.title('Distribuio da Quantidade Vendida')
plt.xlabel('Quantidade Vendida')
plt.ylabel('Frequncia')
plt.grid(True)

# Salvar o grfico como PNG
plt.savefig('distribuicao_quantidade_vendida.png')

plt.show()
```

## Distribuio da Quantidade Vendida

O histograma acima exibe a frequncia da quantidade de produtos vendidos por transao, com uma curva de densidade KDE sobreposta para visualizar a distribuio de forma suavizada.

**Padres Observados:**

* **Distribuio Ampla com Mltiplos Picos:** A quantidade vendida por transao apresenta uma distribuio extensa, variando de valores prximos a zero at aproximadamente 3500 unidades. Notavelmente, a distribuio no possui um nico pico dominante, mas sim mltiplos picos de frequncia, indicando concentraes de vendas em diversas faixas de quantidade.
* **Picos de Frequncia em Faixas Especficas:** Observam-se picos de frequncia em torno de:
    * Valores baixos, mas no o mais proeminente.
    * Aproximadamente 500-600 unidades.
    * Uma faixa mais larga entre 1200 e 1600 unidades, onde se localiza o pico de maior frequncia, embora distribudo.
    * Ao redor de 2600-3000 unidades.
* **Vendas de Mltiplas Unidades como Comportamento Comum:** A ausncia de um pico isolado e massivo na quantidade 1 sugere que as vendas de mltiplas unidades so uma ocorrncia frequente no conjunto de dados.
* **Cauda Direita Substancial:** Embora a frequncia diminua para quantidades muito elevadas, ainda existe um nmero considervel de transaes envolvendo centenas e at milhares de unidades vendidas.
* **Suavizao da Distribuio pela Curva KDE:** A curva de densidade KDE confirma as reas de maior concentrao de vendas identificadas no histograma, suavizando as variaes entre as barras e destacando os mltiplos centros de frequncia.

**Insights e Anlise:**

* **Diversidade nos Tamanhos dos Pedidos:** A presena de mltiplos picos sugere que diferentes tipos de transaes ou clientes podem estar envolvidos. Pode haver vendas no varejo de pequenas quantidades, vendas para pequenos negcios em quantidades mdias e vendas para grandes clientes ou revendedores em quantidades maiores.
* **Identificao de Nveis de Pedido Comuns:** As faixas de quantidade com picos de frequncia (500-600, 1200-1600, 2600-3000) podem representar nveis de pedido tpicos para certos segmentos de clientes ou para promoes especficas que incentivam a compra nessas quantidades.
* **Importncia das Vendas em Volume:** A cauda direita significativa indica que, embora menos frequentes, as vendas de alto volume contribuem de forma importante para o total de unidades vendidas. Analisar essas transaes pode revelar informaes cruciais sobre grandes clientes ou pedidos especiais.
* **Necessidade de Estratgias de Vendas Segmentadas:** Dada a variedade nos tamanhos dos pedidos, pode ser benfico desenvolver estratgias de vendas e marketing segmentadas para atender s necessidades de clientes que compram em diferentes volumes.
* **Investigao dos Motivos por Trs dos Picos:** Investigar os fatores que levam aos picos de frequncia em quantidades especficas pode fornecer insights valiosos sobre o comportamento do cliente, a eficcia de promoes ou a natureza dos produtos mais vendidos em diferentes volumes.dro de vendas.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PrecoVenda', y='VendaQtd', data=df_vendas)
plt.title('Relao entre Preo de Venda e Quantidade')
plt.xlabel('Preo de Venda')
plt.ylabel('Quantidade Vendida')
plt.grid(True)

# Salvar o grfico como PNG
plt.savefig('relacao_preco_venda_quantidade.png')

plt.show()
```

## Relao entre Preo de Venda e Quantidade

Este grfico de disperso explora a relao entre o preo final de venda de um produto e a quantidade desse produto vendida na mesma transao.

**Padres Observados:**

* **Nuvem de Pontos:** A distribuio dos pontos pode indicar se h alguma correlao visual entre as duas variveis.
* **Concentrao:** Pode haver uma maior concentrao de pontos em certas reas do grfico (por exemplo, preos mais baixos com quantidades menores ou vice-versa).
* **Valores Atpicos:** Pontos isolados podem representar vendas incomuns (preos muito altos com quantidades altas/baixas, ou preos muito baixos com quantidades altas/baixas).

**Insights e Anlise:**

* **Correlao Potencial:** Avaliar se h uma tendncia geral. Uma nuvem de pontos com inclinao negativa sugeriria que produtos mais caros tendem a ser comprados em menor quantidade. Uma inclinao positiva sugeriria o contrrio. A ausncia de uma tendncia clara indica pouca correlao direta.
* **Elasticidade de Preo:** Este grfico pode dar uma ideia inicial da elasticidade de preo dos produtos. Se a quantidade vendida diminui significativamente com o aumento do preo, o produto pode ser considerado elstico.
* **Estratgias de Volume por Faixa de Preo:** Analisar as quantidades mdias vendidas em diferentes faixas de preo pode informar estratgias de precificao e promoes direcionadas a faixas especficas.
* **Investigao de Outliers:** Os valores atpicos podem representar promoes especiais, erros de lanamento de preo ou vendas para clientes especficos (como revendedores). Investig-los pode fornecer informaes valiosas.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_vendas['PrecoVenda'], bins=50, kde=True)
plt.title('Distribuio dos Preos de Venda')
plt.xlabel('Preo de Venda')
plt.ylabel('Frequncia')
plt.grid(True)

# Salvar o grfico como PNG
plt.savefig('distribuicao_preco_venda.png')

plt.show()
```

## Distribuio dos Preos de Venda

Este histograma mostra a frequncia dos diferentes preos de venda registrados nas transaes. A curva KDE sobreposta estima a densidade da distribuio dos preos.

**Padres Observados:**

* **Faixa de Preos Mais Comum:** Identificar a faixa de preos com a maior frequncia de vendas. Isso indica a faixa de preo em que a maioria das transaes ocorre.
* **Distribuio (Normal, Assimtrica):** Avaliar se a distribuio dos preos  aproximadamente normal, possui uma assimetria positiva (cauda longa para preos mais altos) ou negativa (cauda longa para preos mais baixos).
* **Picos Mltiplos:** A presena de mltiplos picos pode sugerir diferentes categorias de produtos com faixas de preo distintas.

**Insights e Anlise:**

* **Preo Mdio e Mediana:** A localizao do pico principal (moda) em relao  mdia e  mediana (que voc obteve com `df_vendas.describe()`) pode indicar a forma da distribuio e a presena de valores atpicos.
* **Segmentao de Produtos por Preo:** Se houver mltiplos picos, isso pode indicar segmentos de produtos com diferentes estratgias de preo ou diferentes mercados-alvo.
* **Estratgias de Precificao:** A distribuio dos preos de venda reflete a estratgia de precificao geral da empresa. Analisar essa distribuio ao longo do tempo pode revelar mudanas na estratgia.
* **Comparao com Preo Original:** Comparar essa distribuio com a distribuio dos preos originais pode dar uma ideia do impacto dos descontos.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PrecoOriginal', y='PrecoVenda', data=df_vendas)
plt.plot([df_vendas['PrecoOriginal'].min(), df_vendas['PrecoOriginal'].max()],
         [df_vendas['PrecoOriginal'].min(), df_vendas['PrecoOriginal'].max()],
         'k--', linewidth=2) # Linha de referncia para Preo Original = Preo Venda
plt.title('Relao entre Preo Original e Preo de Venda')
plt.xlabel('Preo Original')
plt.ylabel('Preo de Venda')
plt.grid(True)

# Salvar o grfico como PNG
plt.savefig('relacao_preco_original_venda.png')

plt.show()
```

## Relao entre Preo Original e Preo de Venda

Este grfico de disperso compara o preo original dos produtos com o preo pelo qual eles foram efetivamente vendidos. A linha tracejada diagonal representa o cenrio onde o preo de venda  igual ao preo original (sem desconto).

**Padres Observados:**

* **Pontos Abaixo da Linha:** A maioria dos pontos provavelmente estar abaixo da linha diagonal, indicando que a maioria das vendas ocorreu com algum nvel de desconto.
* **Distncia dos Pontos  Linha:** A distncia vertical de um ponto at a linha diagonal representa a magnitude do desconto aplicado. Pontos mais distantes indicam descontos maiores.
* **Concentrao de Pontos:** Pode haver reas onde os pontos se concentram, sugerindo faixas de desconto mais comuns para certas faixas de preo original.

**Insights e Anlise:**

* **Poltica de Descontos:** A disperso dos pontos em relao  linha diagonal revela a poltica de descontos da empresa. Uma grande concentrao de pontos distantes da linha indica uma estratgia de descontos agressiva.
* **Impacto dos Descontos por Faixa de Preo:** Analisar se a magnitude dos descontos varia com o preo original. Produtos mais caros recebem descontos maiores ou menores?
* **Preo Base vs. Preo Realizado:** Este grfico ajuda a visualizar a diferena entre o preo de tabela e o preo real pelo qual os produtos so vendidos.
* **Eficcia dos Descontos:** Relacionar a profundidade dos descontos com o volume de vendas (observado no grfico anterior) pode ajudar a avaliar a eficcia da estratgia de descontos em impulsionar as vendas.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_vendas['Desconto'], bins=30, kde=True)
plt.title('Distribuio do Valor do Desconto')
plt.xlabel('Valor do Desconto')
plt.ylabel('Frequncia')
plt.grid(True)

# Salvar o grfico como PNG
plt.savefig('distribuicao_valor_desconto.png')

plt.show()
```

* **Valores Atpicos:** Descontos com valores muito altos podem representar promoes especiais ou erros.
