```python
import pandas as pd
```

```python
df_vendas = pd.read_excel('dadosVenda.xlsx')
df_vendas.head(3)
```

```python
df_vendas.info()
```

```python
df_vendas.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df_vendas['VendaQtd'], bins=30, kde=True)
plt.title('Distribuição da Quantidade Vendida')
plt.xlabel('Quantidade Vendida')
plt.ylabel('Frequência')
plt.grid(True)

# Salvar o gráfico como PNG
plt.savefig('distribuicao_quantidade_vendida.png')

plt.show()
```

## Distribuição da Quantidade Vendida

O histograma acima exibe a frequência da quantidade de produtos vendidos por transação, com uma curva de densidade KDE sobreposta para visualizar a distribuição de forma suavizada.

**Padrões Observados:**

* **Distribuição Ampla com Múltiplos Picos:** A quantidade vendida por transação apresenta uma distribuição extensa, variando de valores próximos a zero até aproximadamente 3500 unidades. Notavelmente, a distribuição não possui um único pico dominante, mas sim múltiplos picos de frequência, indicando concentrações de vendas em diversas faixas de quantidade.
* **Picos de Frequência em Faixas Específicas:** Observam-se picos de frequência em torno de:
    * Valores baixos, mas não o mais proeminente.
    * Aproximadamente 500-600 unidades.
    * Uma faixa mais larga entre 1200 e 1600 unidades, onde se localiza o pico de maior frequência, embora distribuído.
    * Ao redor de 2600-3000 unidades.
* **Vendas de Múltiplas Unidades como Comportamento Comum:** A ausência de um pico isolado e massivo na quantidade 1 sugere que as vendas de múltiplas unidades são uma ocorrência frequente no conjunto de dados.
* **Cauda Direita Substancial:** Embora a frequência diminua para quantidades muito elevadas, ainda existe um número considerável de transações envolvendo centenas e até milhares de unidades vendidas.
* **Suavização da Distribuição pela Curva KDE:** A curva de densidade KDE confirma as áreas de maior concentração de vendas identificadas no histograma, suavizando as variações entre as barras e destacando os múltiplos centros de frequência.

**Insights e Análise:**

* **Diversidade nos Tamanhos dos Pedidos:** A presença de múltiplos picos sugere que diferentes tipos de transações ou clientes podem estar envolvidos. Pode haver vendas no varejo de pequenas quantidades, vendas para pequenos negócios em quantidades médias e vendas para grandes clientes ou revendedores em quantidades maiores.
* **Identificação de Níveis de Pedido Comuns:** As faixas de quantidade com picos de frequência (500-600, 1200-1600, 2600-3000) podem representar níveis de pedido típicos para certos segmentos de clientes ou para promoções específicas que incentivam a compra nessas quantidades.
* **Importância das Vendas em Volume:** A cauda direita significativa indica que, embora menos frequentes, as vendas de alto volume contribuem de forma importante para o total de unidades vendidas. Analisar essas transações pode revelar informações cruciais sobre grandes clientes ou pedidos especiais.
* **Necessidade de Estratégias de Vendas Segmentadas:** Dada a variedade nos tamanhos dos pedidos, pode ser benéfico desenvolver estratégias de vendas e marketing segmentadas para atender às necessidades de clientes que compram em diferentes volumes.
* **Investigação dos Motivos por Trás dos Picos:** Investigar os fatores que levam aos picos de frequência em quantidades específicas pode fornecer insights valiosos sobre o comportamento do cliente, a eficácia de promoções ou a natureza dos produtos mais vendidos em diferentes volumes.drão de vendas.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PrecoVenda', y='VendaQtd', data=df_vendas)
plt.title('Relação entre Preço de Venda e Quantidade')
plt.xlabel('Preço de Venda')
plt.ylabel('Quantidade Vendida')
plt.grid(True)

# Salvar o gráfico como PNG
plt.savefig('relacao_preco_venda_quantidade.png')

plt.show()
```

## Relação entre Preço de Venda e Quantidade

Este gráfico de dispersão explora a relação entre o preço final de venda de um produto e a quantidade desse produto vendida na mesma transação.

**Padrões Observados:**

* **Nuvem de Pontos:** A distribuição dos pontos pode indicar se há alguma correlação visual entre as duas variáveis.
* **Concentração:** Pode haver uma maior concentração de pontos em certas áreas do gráfico (por exemplo, preços mais baixos com quantidades menores ou vice-versa).
* **Valores Atípicos:** Pontos isolados podem representar vendas incomuns (preços muito altos com quantidades altas/baixas, ou preços muito baixos com quantidades altas/baixas).

**Insights e Análise:**

* **Correlação Potencial:** Avaliar se há uma tendência geral. Uma nuvem de pontos com inclinação negativa sugeriria que produtos mais caros tendem a ser comprados em menor quantidade. Uma inclinação positiva sugeriria o contrário. A ausência de uma tendência clara indica pouca correlação direta.
* **Elasticidade de Preço:** Este gráfico pode dar uma ideia inicial da elasticidade de preço dos produtos. Se a quantidade vendida diminui significativamente com o aumento do preço, o produto pode ser considerado elástico.
* **Estratégias de Volume por Faixa de Preço:** Analisar as quantidades médias vendidas em diferentes faixas de preço pode informar estratégias de precificação e promoções direcionadas a faixas específicas.
* **Investigação de Outliers:** Os valores atípicos podem representar promoções especiais, erros de lançamento de preço ou vendas para clientes específicos (como revendedores). Investigá-los pode fornecer informações valiosas.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_vendas['PrecoVenda'], bins=50, kde=True)
plt.title('Distribuição dos Preços de Venda')
plt.xlabel('Preço de Venda')
plt.ylabel('Frequência')
plt.grid(True)

# Salvar o gráfico como PNG
plt.savefig('distribuicao_preco_venda.png')

plt.show()
```

## Distribuição dos Preços de Venda

Este histograma mostra a frequência dos diferentes preços de venda registrados nas transações. A curva KDE sobreposta estima a densidade da distribuição dos preços.

**Padrões Observados:**

* **Faixa de Preços Mais Comum:** Identificar a faixa de preços com a maior frequência de vendas. Isso indica a faixa de preço em que a maioria das transações ocorre.
* **Distribuição (Normal, Assimétrica):** Avaliar se a distribuição dos preços é aproximadamente normal, possui uma assimetria positiva (cauda longa para preços mais altos) ou negativa (cauda longa para preços mais baixos).
* **Picos Múltiplos:** A presença de múltiplos picos pode sugerir diferentes categorias de produtos com faixas de preço distintas.

**Insights e Análise:**

* **Preço Médio e Mediana:** A localização do pico principal (moda) em relação à média e à mediana (que você obteve com `df_vendas.describe()`) pode indicar a forma da distribuição e a presença de valores atípicos.
* **Segmentação de Produtos por Preço:** Se houver múltiplos picos, isso pode indicar segmentos de produtos com diferentes estratégias de preço ou diferentes mercados-alvo.
* **Estratégias de Precificação:** A distribuição dos preços de venda reflete a estratégia de precificação geral da empresa. Analisar essa distribuição ao longo do tempo pode revelar mudanças na estratégia.
* **Comparação com Preço Original:** Comparar essa distribuição com a distribuição dos preços originais pode dar uma ideia do impacto dos descontos.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PrecoOriginal', y='PrecoVenda', data=df_vendas)
plt.plot([df_vendas['PrecoOriginal'].min(), df_vendas['PrecoOriginal'].max()],
         [df_vendas['PrecoOriginal'].min(), df_vendas['PrecoOriginal'].max()],
         'k--', linewidth=2) # Linha de referência para Preço Original = Preço Venda
plt.title('Relação entre Preço Original e Preço de Venda')
plt.xlabel('Preço Original')
plt.ylabel('Preço de Venda')
plt.grid(True)

# Salvar o gráfico como PNG
plt.savefig('relacao_preco_original_venda.png')

plt.show()
```

## Relação entre Preço Original e Preço de Venda

Este gráfico de dispersão compara o preço original dos produtos com o preço pelo qual eles foram efetivamente vendidos. A linha tracejada diagonal representa o cenário onde o preço de venda é igual ao preço original (sem desconto).

**Padrões Observados:**

* **Pontos Abaixo da Linha:** A maioria dos pontos provavelmente estará abaixo da linha diagonal, indicando que a maioria das vendas ocorreu com algum nível de desconto.
* **Distância dos Pontos à Linha:** A distância vertical de um ponto até a linha diagonal representa a magnitude do desconto aplicado. Pontos mais distantes indicam descontos maiores.
* **Concentração de Pontos:** Pode haver áreas onde os pontos se concentram, sugerindo faixas de desconto mais comuns para certas faixas de preço original.

**Insights e Análise:**

* **Política de Descontos:** A dispersão dos pontos em relação à linha diagonal revela a política de descontos da empresa. Uma grande concentração de pontos distantes da linha indica uma estratégia de descontos agressiva.
* **Impacto dos Descontos por Faixa de Preço:** Analisar se a magnitude dos descontos varia com o preço original. Produtos mais caros recebem descontos maiores ou menores?
* **Preço Base vs. Preço Realizado:** Este gráfico ajuda a visualizar a diferença entre o preço de tabela e o preço real pelo qual os produtos são vendidos.
* **Eficácia dos Descontos:** Relacionar a profundidade dos descontos com o volume de vendas (observado no gráfico anterior) pode ajudar a avaliar a eficácia da estratégia de descontos em impulsionar as vendas.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_vendas['Desconto'], bins=30, kde=True)
plt.title('Distribuição do Valor do Desconto')
plt.xlabel('Valor do Desconto')
plt.ylabel('Frequência')
plt.grid(True)

# Salvar o gráfico como PNG
plt.savefig('distribuicao_valor_desconto.png')

plt.show()
```

## Distribuição do Valor do Desconto

Este histograma exibe a frequência dos diferentes valores de desconto aplicados nas vendas. A curva KDE sobreposta fornece uma estimativa da densidade da distribuição dos valores de desconto.

**Padrões Observados:**

* **Faixa de Desconto Mais Comum:** Identificar a faixa de valores de desconto que ocorre com maior frequência.
* **Distribuição:** Avaliar a forma da distribuição (simétrica, assimétrica, etc.).
* **Valores Atípicos:** Descontos com valores muito altos podem representar promoções especiais ou erros.

**Insights e Análise:**

* **Valor Típico de Desconto:** O pico do histograma indica o valor de desconto mais comum oferecido.
* **Impacto Financeiro dos Descontos:** Analisar a distribuição dos valores de desconto pode ajudar a entender o impacto financeiro total dos descontos nas receitas.
* **Estratégias de Desconto:** A distribuição pode revelar se os descontos são geralmente pequenos e frequentes ou grandes e ocasionais.
* **Necessidade de Padronização:** Uma distribuição muito dispersa pode sugerir a necessidade de revisar e padronizar a política de descontos.
